<b>Práctica de laboratorio: Comparar datos con un hash</b>

Paso 1:	Crear un archivo de texto
<hr>
<br>
a.	Busque en su equipo el programa Bloc de notas y ábralo.
<br>
b.	Escriba algún texto en el programa.
<br>

<img src="img/imagen1.png">
<br>
c.	Elija Archivo > Guardar.
<br>
d.	Navegue hasta Escritorio.
<br>
e.	Escriba Hash en el campo Nombre de archivo: y haga clic en Guardar.




<b>Paso 2:	Instalar HashCalc </b><br>
<hr>
a.	Abra un navegador web y vaya a http://www.slavasoft.com/download.htm.
<br>
<img src="img/imagen2.png">
<br>
b.	Haga clic en Descargar en la fila HashCalc 2.02.
<br>
c.	Abra el archivo hashcalc.zip y ejecute el archivo setup.exe que está en su interior.
<br>
<img src="img/imagen3.png">
<br>
d.	Siga el asistente de instalación para instalar HashCalc.
<br>
e.	Haga clic en Finalizar en la última pantalla y cierre el archivo README si está abierto. Puede leer el archivo, si lo desea.
<br>f.	HashCalc ahora está instalado y en funcionamiento.
<br><img src="img/imagen4.png">

<b>Paso 3:	Calcular un hash del archivo Hash.txt </b>
<hr>
<br>a.	Establezca los siguientes elementos en HashCalc:
<br>1)	Formato de los datos: Archivo.
<br>2)	Datos: haga clic en el botón … que está junto al campo Datos, navegue hasta el Escritorio y elija el archivo Hash.txt.
<br>3)	Quite la selección de HMAC.
<br>4)	Quite la selección de todos los tipos de hash, excepto MD5.
<br>
b.	Haga clic en el botón Calcular.
<br>¿Cuál es el valor junto a MD5? <b>c50618d14b2baa1a69009c44d58901ed</b>
<br><img src="img/imagen5.png">


<b>Paso 4:	Haga un cambio en el archivo Hash.txt</b>
<br>a.	Navegue hasta el Escritorio y abra el archivo Hash.txt.
<br>b.	Realice un cambio menor en el texto, como eliminar una letra, o agregar un espacio o un punto.
<br>c.	Haga clic en Archivo > Guardar y cierre el Bloc de notas.
<br><hr><b>Paso 5:	Calcule un nuevo hash del archivo Hash.txt</b>
<br>a.	Haga clic en el botón Calcular en HashCalc nuevamente.
<br>¿Cuál es el valor junto a MD5? <b>06ecc17437eece74ce327678d039cbe5</b>
<br>
<img src="img/imagen6.png">
<br>¿El valor es diferente del valor registrado en el paso 3?
<br>
<b>Si, el valor del hash code ha cambiado</b>
<br>b.	Coloque una marca junto a todos los tipos de hash.
<br>c.	Haga clic en Calcular.
<br>d.	Fíjese cómo muchos de los tipos de hash crean un hash de longitud diferente. ¿Por qué? Cada decodificacion usa diferentes métodos de encriptacion
<br><b>Cada uno tiene su propio algoritmo de encriptamiento y por eso el resultado con cada uno es diferente</b>
